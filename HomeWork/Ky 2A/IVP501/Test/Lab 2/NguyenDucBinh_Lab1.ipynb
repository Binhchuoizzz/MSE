```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# ========== 1. Color Balance ==========


def color_balance(img, r_gain=1.0, g_gain=1.0, b_gain=1.0):
    b, g, r = cv2.split(img)
    r = cv2.multiply(r, r_gain)
    g = cv2.multiply(g, g_gain)
    b = cv2.multiply(b, b_gain)
    return cv2.merge([b, g, r])

# ========== 2. Histogram Equalization ==========


def histogram_equalization(img_gray):
    return cv2.equalizeHist(img_gray)

# ========== 3. Median Filter ==========


def apply_median_filter(img):
    return cv2.medianBlur(img, 3)

# ========== 4. Mean Filter ==========


def apply_mean_filter(img):
    return cv2.blur(img, (3, 3))

# ========== 5. Gaussian Smoothing ==========


def apply_gaussian_smoothing(img):
    return cv2.GaussianBlur(img, (5, 5), 1)

# ========== Main Program ==========


def process_image(image_path):
    # Load original image
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Cannot load image from {image_path}")

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply all processing steps
    balanced = color_balance(img, 1.2, 1.0, 0.8)
    equalized = histogram_equalization(gray)
    median = apply_median_filter(gray)
    mean = apply_mean_filter(gray)
    gaussian = apply_gaussian_smoothing(gray)

    # Plot histograms
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 3, 1)
    plt.title("Original Histogram")
    plt.hist(gray.ravel(), 256, [0, 256])

    plt.subplot(2, 3, 2)
    plt.title("Equalized Histogram")
    plt.hist(equalized.ravel(), 256, [0, 256])

    # Show images
    cv2.imshow("Original", img)
    cv2.imshow("Color Balanced", balanced)
    cv2.imshow("Equalized", equalized)
    cv2.imshow("Median Filter", median)
    cv2.imshow("Mean Filter", mean)
    cv2.imshow("Gaussian Smoothing", gaussian)

    plt.tight_layout()
    plt.show()
    cv2.waitKey(0)
    cv2.destroyAllWindows()


def color_balance(img, r_gain=1.0, g_gain=1.0, b_gain=1.0):
    b, g, r = cv2.split(img)
    r = cv2.multiply(r, r_gain)
    g = cv2.multiply(g, g_gain)
    b = cv2.multiply(b, b_gain)
    return cv2.merge([b, g, r])


def histogram_equalization(img_gray):
    return cv2.equalizeHist(img_gray)


def apply_median_filter(img):
    return cv2.medianBlur(img, 3)


def apply_mean_filter(img):
    return cv2.blur(img, (3, 3))


def apply_gaussian_smoothing(img):
    return cv2.GaussianBlur(img, (5, 5), 1)


def process_image(image_path):
    img = cv2.imread(image_path)
    if img is None:
        raise FileNotFoundError(f"Cannot load image from {image_path}")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    balanced = color_balance(img, 1.2, 1.0, 0.8)
    equalized = histogram_equalization(gray)
    median = apply_median_filter(gray)
    mean = apply_mean_filter(gray)
    gaussian = apply_gaussian_smoothing(gray)

    plt.figure(figsize=(12, 6))
    plt.subplot(2, 3, 1)
    plt.title("Original Histogram")
    plt.hist(gray.ravel(), 256, [0, 256])

    plt.subplot(2, 3, 2)
    plt.title("Equalized Histogram")
    plt.hist(equalized.ravel(), 256, [0, 256])

    cv2.imshow("Original", img)
    cv2.imshow("Color Balanced", balanced)
    cv2.imshow("Equalized", equalized)
    cv2.imshow("Median Filter", median)
    cv2.imshow("Mean Filter", mean)
    cv2.imshow("Gaussian Smoothing", gaussian)

    plt.tight_layout()
    plt.show()
    cv2.waitKey(0)
    cv2.destroyAllWindows()


# Example usage:
process_image(
    'E:\\MSE\\HomeWork\\Ky 2\\IVP501\\Lab & Test & Project\Lab 2\\717f5817-8408-429c-8328-81b71478ec27.jpg')
```

```python
import numpy as np
import cv2
image = np.ones((1080, 1920, 3), dtype=np.uint8)*255
cv2.imshow('White background', img)
# cv2.imwrite('White bg.png',img)
cv2.waitKey()
```

```python
def draw_rectangle(event, x, y, flags, param):
    global rect_start, rect_end, drawing, image, rect_coords
    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        rect_start = (x, y)
    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing:
            image_copy = image.copy()
            cv2.rectangle(image_copy, rect_start, (x, y), (0, 255, 0), 2)
            cv2.imshow('Draw rectangle', image_copy)

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        rect_end = (x, y)
        cv2.rectangle(image_copy, rect_start, rect_end, (0, 255, 0), 2)
        cv2.imshow('Draw rectangle', image)
```

```python
cv2.namedWindow('Draw rectangle')
cv2.setMouseCallback('Draw rectangle', draw_rectangle)
cv2.waitKey(0)
```

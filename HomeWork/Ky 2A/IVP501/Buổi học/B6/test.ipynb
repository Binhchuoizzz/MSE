```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans  # <- sửa ở đây
```

```python
img = plt.imread(r'D:\MSE\HomeWork\Ky 2\IVP501\Test\IMG\329cea444b7efe20a76f1.jpg')
x = np.reshape(img,(-1,3))
print(x)
plt.imshow(img)
```

```python
kmean = KMeans(5)
```

```python
kmean.fit(x)
```

```python
clusters = kmean.cluster_centers_.astype('uint8')
labels = kmean.labels_
# print(clusters)
print(labels)
```

```python
seg_x = clusters[labels]
img_seg =seg_x.reshape(img.shape)
print(img_seg)
```

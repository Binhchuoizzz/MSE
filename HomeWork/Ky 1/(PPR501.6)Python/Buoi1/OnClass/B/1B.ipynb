```python
import pandas as pd

AAPL = pd.read_csv(path_data + 'AAPL.csv')
GOOG = pd.read_csv(path_data + 'GOOG.csv')
MSFT = pd.read_csv(path_data + 'MSFT.csv')
print(type(AAPL), type(GOOG), type(MSFT))

print('Shape AAPL : ', AAPL.shape, ' Rows:',
      AAPL.shape[0], ' Cols:', AAPL.shape[1])

print('Shape GOOG : ', GOOG.shape, ' Rows:',
      GOOG.shape[0], ' Cols:', GOOG.shape[1])

print('Shape MSFT : ', MSFT.shape, ' Rows:',
      MSFT.shape[0], ' Cols:', MSFT.shape[1])
```

```python
display(AAPL.head())
display(GOOG.tail(10))
display(MSFT.head(3))
display(MSFT.iloc[2:4])
display(MSFT.loc[2:4])
```

```python
AAPL['Adj Close'] = AAPL['Adj Close'].round(decimals=2)
AAPL['Adj Close'].head()
```

```python
import statistics as s

adjValues = AAPL['Adj Close'].values
# print(adjValues)
mean = s.mean(adjValues)

median = s.median(adjValues)

mode = s.mode(adjValues)

std = s.stdev(adjValues)

print(f"mean = {mean} and std = {std} and mode = {mode} and median = {median}")
```

```python
import seaborn as sns
sns.set_style('dark')
AAPL['Adj Close'].plot(legend=True, figsize=(12, 8))
```

```python
ma = 10
column_name = "MA %s Days" % (str(ma))
AAPL[column_name] = AAPL['Adj Close'].rolling(window=ma, center=False).mean()

ma = 20
column_name = "MA %s Days" % (str(ma))
AAPL[column_name] = AAPL['Adj Close'].rolling(window=ma, center=False).mean()

ma = 30
column_name = "MA %s Days" % (str(ma))
AAPL[column_name] = AAPL['Adj Close'].rolling(window=ma, center=False).mean()
```

```python
# In order calculate daily return we can use the percentage change of the adjusted closing price
AAPL['Percent Change'] = AAPL['Adj Close'].pct_change()
AAPL.head()
```

```python
# Let us now plot the daily return
AAPL['Percent Change'].plot(figsize=(12, 5))
```

```python
all_stocks = pd.DataFrame(
    columns=['Apple', 'Google', 'Microsoft'], index=AAPL['Date'])
all_stocks['Apple'] = AAPL['Adj Close'].values

all_stocks['Google'] = GOOG['Adj Close'].values

all_stocks['Microsoft'] = MSFT['Adj Close'].values

all_stocks.tail()
```

```python
import numpy as np
corr = all_stocks.corr()
print(corr)
sns.heatmap(corr, annot=True, mask=False)
```

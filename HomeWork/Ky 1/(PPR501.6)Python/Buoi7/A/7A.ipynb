```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
# path = 'D:/Installed/Fonts/'
# namefont = ['iOS-Helvetica.ttf', 'Tahoma.ttf',
#             'iOS-Helvetica-Light.ttf', 'IOS7HelveticaNeue-CondensedBlack.ttf']
# fpath = path+namefont[2]
# fontprop = fm.FontProperties(fname=fpath)
# Creating the dataset
df = sns.load_dataset('titanic')
df = df.groupby('who')['fare'].sum().to_frame().reset_index()

# Creating the bar chart
plt.barh(df['who'], df['fare'], color=[
         (85/255, 85/255, 127/255), '#E6E6FA', '#B0E0E6'], label='people')
plt.grid(color='gray', linestyle='-', linewidth=2, which='both', alpha=0.8)
# Adding the aesthetics
plt.title('Chart title')
plt.xlabel('X axis title', fontweight='bold',
           fontsize=18)
plt.ylabel('Y axis title', fontweight='bold', fontsize=18)
plt.xticks(fontsize=12, rotation=70)
plt.yticks(fontsize=10, rotation=90)

plt.xlim(0, 16000)
legend = plt.legend(loc=1,  # bbox_to_anchorloc=[0.1,0.35, 0.95, 0.65],
                    ncol=2,  borderaxespad=0.6,  labelspacing=1, edgecolor='black', framealpha=0.6),  # ,mode="expand
plt.grid(False)
# plt.axis('off')
# Show the plot
# dir_out = 'G:/My Drive/VTCA/Hieu/less_7/'
# plt.savefig(dir_out+'bar.png', dpi=600, pad_inches=0.01)
plt.show()
```

```python
df1 = sns.load_dataset('titanic')
df1 = df1.groupby('who')['fare'].sum().to_frame().reset_index()
plt.bar(df1['who'], df1['fare'], label='person')
sns.barplot(x='who', y='fare', data=df1, palette='Blues')
plt.title('Chart title')
plt.xlabel('X axis title', fontweight='bold',
           fontsize=18)
plt.ylabel('Y axis title', fontweight='bold', fontsize=18)
plt.show()
```

```python
import pandas as pd
df1 = sns.load_dataset('titanic')
# df1.head()
df_pivot = pd.pivot_table(df1, values='fare', index='who', columns='class')

df_pivot.plot(kind='bar', alpha=0.5)

plt.title('Chart title')
plt.xlabel('X axis title', fontweight='bold',
           fontsize=18)
plt.ylabel('Y axis title', fontweight='bold', fontsize=18)
plt.show()
```

```python
# Stacked bar chart
# Creating the dataset
df = pd.DataFrame(columns=["A", "B", "C", "D"],
                  data=[["E", 3, 1, 1],
                        ["F", 2, 1, 0],
                        ["G", 0, 1, 0]])

df.plot.bar(x='A', y=["B", "C", "D"],  stacked=True,  width=0.4, alpha=0.5)

# Adding the aesthetics
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')

# Show the plot
plt.show()
```

```python
dataframe = pd.DataFrame(columns=["A", "B", "C", "D"],
                         data=[["E", 0, 1, 1],
                               ["F", 1, 1, 0],
                               ["G", 0, 1, 0]])
dataframe.set_index('A').T.plot(kind='bar', stacked=True)
# Adding the aesthetics
plt.title('Chart title')
plt.xlabel('X axis title')
plt.ylabel('Y axis title')
# Show the plot
plt.show()
```

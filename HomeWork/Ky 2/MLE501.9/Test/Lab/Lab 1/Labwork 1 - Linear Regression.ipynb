```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Training data
# y score - x - studying hours # MSE22
X_train = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y_train = np.array([2, 4, 5, 4, 5])

# Testing data #MSE23
X_test = np.array([3, 4, 5]).reshape(-1, 1)

# predict the score based on x-test
# grouth-truth
Y_test = np.array([4, 4, 4.5]).reshape(-1, 1)

# Fit model on training data
model = LinearRegression()
model.fit(X_train, Y_train)

# Predict values on training and test sets
Y_train_pred = model.predict(X_train)
Y_test_pred = model.predict(X_test)

# Plotting
plt.scatter(X_train, Y_train, color='green', label='Training Data')
plt.scatter(X_test, Y_test, color='blue', label='Testing Data')

# Regression line over the full X range
X_all = np.vstack((X_train, X_test))
X_line = np.linspace(X_all.min(), X_all.max(), 100).reshape(-1, 1)
Y_line = model.predict(X_line)
plt.plot(X_line, Y_line, color='red', label='Regression Line')

plt.xlabel('Hours Studied')
plt.ylabel('Score')
plt.title('Linear Regression: Training vs Testing')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Set random seed for reproducibility
np.random.seed(42)

# Generate sample data
n_samples = 15
hours_studied = np.random.randint(10, 31, size=n_samples)  # 10 to 30
practice_tests = np.random.randint(0, 6, size=n_samples)    # 0 to 5

# Create target scores using a formula + noise
# Assume: score = 0.3 * hours + 1.0 * practice_tests + noise
noise = np.random.normal(0, 1.5, n_samples)
scores = 0.3 * hours_studied + 1.0 * practice_tests + noise

# Clip scores to range [0, 10]
scores = np.clip(scores, 0, 10)

# Prepare feature matrix
X = np.column_stack((hours_studied, practice_tests))
Y = scores

# Train linear regression model
model = LinearRegression()
model.fit(X, Y)

# Output equation
print(f"Regression Equation: Y = {model.coef_[0]:.2f}*X1 + {model.coef_[1]:.2f}*X2 + {model.intercept_:.2f}")

# Predict and visualize
Y_pred = model.predict(X)

# Plot actual vs predicted scores
plt.scatter(range(n_samples), Y, color='blue', label='Actual Score')
plt.plot(range(n_samples), Y_pred, color='red', linestyle='--', label='Predicted Score')
plt.xlabel('Sample Index')
plt.ylabel('Score (0–10)')
plt.title('Multivariate Linear Regression on Simulated Student Data')
plt.legend()
plt.grid(True)
plt.show()
```

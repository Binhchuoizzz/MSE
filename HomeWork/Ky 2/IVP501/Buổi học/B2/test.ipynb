```python
# tao hinh cn
def draw_rectangle(event, x,y , flags, param):
     global rect_start, rect_end, drawing, image, rect_coords
     if event == cv2.EVENT_LBUTTONDBLCLK: # di chuột
          drawing = True
          rect_stat = (x,y)
     elif event == cv2.EVENT_MOUSEMOVE: # chuột di
          if drawing:
               image_copy = image.copy()
               cv2.rectangle(image_copy, rect_stat, (x,y), (0,255,0),2)
     elif event == cv2.EVENT_LBUTTONUP: # thả chuột
          drawing = False
          rect_end = (x,y)
          cv2.rectangle(image_copy, rect_stat, (x,y), (0,255,0),2)
          cv2.imshow('Draw Rectangle', image)
          
```

```python
cv2.namedWindow('Draw Rectangle')
cv2.setMouseCallback('Draw Rectangle', draw_rectangle)
cv2.waitKey()
```

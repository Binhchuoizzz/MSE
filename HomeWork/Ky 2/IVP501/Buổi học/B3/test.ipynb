```python
img4 = cv2.imread("D:\MSE\HomeWork\Ky 2\IVP501\Book\images\Equalization.png",0)
cv2.imshow('original',img4)
cv2.waitKey(0)
```

```python
print(img4)
```

```python
hist = cv2.calcHist(img4,[0],None,[256],[0,256])
```

```python
plt.plot(hist)
```

```python
cumulater = np.zeros_like(hist,np.float64)
for i in range(len(hist)):
     cumulater[i] =  hist[i].sum()
     
new_hist = (cumulater - cumulater.min())/(cumulater.max()-cumulater.min())*255
```

```python
plt.plot(hist)
plt.plot(new_hist)
```

```python
plt.plot(new_hist)

```

```python
plt.plot(cumulater)
```

```python
img_equal = img4.copy()
```

```python
h,w = img4.shape
for i in range(h):
     for j in range(w):
          img_equal[i,j]= new_hist[img_equal[i,j]]
          
```

```python
cv2.imshow('equal',img_equal)    
cv2.imshow('origin',img4)
cv2.waitKey(0)
```

```python
import cv2
import numpy as np

# 1. Đọc ảnh và chuyển sang grayscale
img = cv2.imread(r'D:\MSE\HomeWork\Ky 2\IVP501\Book\images\foto1B.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

# 2. Tính Harris response
# blockSize: kích thước vùng lân cận
# ksize: kích thước kernel Sobel
# k: hệ số Harris (thường 0.04–0.06)
dst = cv2.cornerHarris(src=gray,
                       blockSize=2,
                       ksize=3,
                       k=0.04)

# 3. Nở rộng góc để dễ nhìn
dst = cv2.dilate(dst, None)

# 4. Threshold để xác định corner
# Mọi pixel có giá trị > 1% max(dst) sẽ được đánh dấu
threshold = 0.01 * dst.max()
corner_img = img.copy()
corner_img[dst > threshold] = [0, 0, 255]  # tô đỏ

# 5. Hiển thị kết quả
cv2.imshow('Harris Corners', corner_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

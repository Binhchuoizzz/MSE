✅ 1. Translation (Tịnh tiến)


```python
import numpy as np
import matplotlib.pyplot as plt

# Điểm gốc
p = np.array([2, 3])
# Vector tịnh tiến
t = np.array([3, 2])
# Điểm sau khi dịch
p_new = p + t

plt.scatter(*p, color='blue', label='Original')
plt.scatter(*p_new, color='red', label='Translated')
plt.legend()
plt.grid()
plt.title('Translation')
plt.show()
```

✅ 2. Rotation (Xoay quanh gốc)


```python
theta = np.radians(45)  # 45 độ
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])

p = np.array([2, 1])
p_new = R @ p

plt.scatter(*p, color='blue', label='Original')
plt.scatter(*p_new, color='red', label='Rotated')
plt.legend()
plt.grid()
plt.title('Rotation')
plt.axis('equal')
plt.show()
```

✅ 3. Scaling (Tỷ lệ)


```python
S = np.array([[2, 0],
              [0, 1.5]])  # Tăng x gấp đôi, y gấp 1.5

p = np.array([2, 1])
p_new = S @ p

plt.scatter(*p, color='blue', label='Original')
plt.scatter(*p_new, color='red', label='Scaled')
plt.legend()
plt.grid()
plt.title('Scaling')
plt.axis('equal')
plt.show()
```

✅ 4. Affine Transformation


```python
A = np.array([[1, 0.5, 1],  # xoay nhẹ + tịnh tiến
              [0.2, 1, 2]])

p = np.array([2, 1, 1])  # điểm dạng thuần nhất (x, y, 1)
p_new = A @ p

plt.scatter(p[0], p[1], color='blue', label='Original')
plt.scatter(p_new[0], p_new[1], color='red', label='Affine')
plt.legend()
plt.grid()
plt.title('Affine Transformation')
plt.axis('equal')
plt.show()
```

✅ 5. Projective Transformation (Biến đổi phối cảnh)


```python
H = np.array([[1, 0.2, 0],
              [0.1, 1, 0],
              [0.001, 0.002, 1]])

p = np.array([2, 1, 1])  # dạng thuần nhất
p_proj = H @ p
p_proj = p_proj / p_proj[2]  # chuẩn hóa lại tọa độ

plt.scatter(p[0], p[1], color='blue', label='Original')
plt.scatter(p_proj[0], p_proj[1], color='red', label='Projective')
plt.legend()
plt.grid()
plt.title('Projective Transformation')
plt.axis('equal')
plt.show()
```

Loại - Biến dạng hình ban đầu thành...
Translation - Hình y hệt, chỉ bị dời đi
Rigid - Xoay hình nhưng giữ kích thước
Similarity - Phóng to / thu nhỏ và xoay, vẫn đúng hình dạng
Affine - Bóp méo hình một chút (shear)
Projective - Méo theo phối cảnh (như nhìn từ góc xiên)

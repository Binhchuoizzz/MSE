```python
# Function
def printscreen(s = None):
  "This function is to print str to screen"
  if(s): # s is None
    print(s)
  else:
    print("This is the default call a function")
  return

printscreen("This is the first call a function")
s = "This is the second call a function"

printscreen(s)
printscreen()
```
